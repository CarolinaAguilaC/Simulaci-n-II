# Estimación del número $\pi$

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

Partimos de un experimento que puede describirse como el lanzamiento al azar de un dardo.

Los resultados posibles son:

a) Cae dentro del cuarto de círculo
b) Cae fuera del cuarto de círculo

Y construimos un código para simular su repetición utilizando la generación de números aleatorios (uniformemente distribuidos).

```python
N = 1000
Aciertos = 0

XCircle=[]
YCircle=[]
XSquare=[]
YSquare=[]

for p in range(N):
    x=random.random()
    y=random.random()
    if(x**2+y**2 <= 1):
        Aciertos+=1
        XCircle.append(x)
        YCircle.append(y)
    else:
        XSquare.append(x)
        YSquare.append(y)

Pi = 4*Aciertos/N

print("N=%d Aciertos=%d Pi=%.2f" %(N,Aciertos,Pi))

XLin=np.linspace(0,1)
YLin=[]
for x in XLin:
    YLin.append(np.sqrt(1-x**2))

fig = plt.figure(1, figsize=(8, 8))
plt.axis("equal")
plt.grid(which="major")
plt.plot(XLin , YLin, color="red" , linewidth="1")
plt.scatter(XCircle, YCircle, color="blue", marker   =".")
plt.scatter(XSquare, YSquare, color="green"  , marker   =".")
plt.title("Estimación de $\pi$ por el método de Monte Carlo")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

Visualizamos el comportamiento de nuestra estimación en función del número de iteraciones.

```python
def pi_mc(n):
    pi = 0
    lx = []
    ly = []
    counter = 0
    for i in range(n):
        x = random.random()
        y = random.random()
        if x**2 + y**2 <1:
            counter = counter + 1
            lx.append(counter)
            ly.append(4.0*float(counter)/(i + 1))

        pi = 4.0*float(counter)/(i + 1)

        e = np.absolute((pi - np.pi)/np.pi)
    print("pi =", pi)
    print("error =", e)

    fig = plt.figure(1, figsize=(8, 8))
    plt.xlabel("Iteraciones")
    plt.ylabel("Estimación de $\pi$")
    plt.grid()
    plt.plot(lx, ly, color="blue", markersize=1)
    plt.show()
```

```python
pi_mc(1000)
```

```python
import random
import math

# Parámetros del experimento
l = 1  # Longitud de la aguja
d = 2  # Espacio entre las líneas paralelas
n = 100000  # Número de lanzamientos de la aguja

# Contador de intersecciones
intersecciones = 0

# Simulación del experimento
for _ in range(n):
    # Generar posición y ángulo aleatorios
    x = random.uniform(0, d)
    theta = random.uniform(0, math.pi/2)

    # Calcular posición de la punta de la aguja
    x_punta = x + l * math.cos(theta)

    # Verificar si la aguja intersecta una línea
    if x_punta >= d:
        intersecciones += 1

# Estimación de π utilizando la fórmula de Buffon
estimacion_pi = (2 * l * n) / (d * intersecciones)

print("Estimación de π:", estimacion_pi)
```
