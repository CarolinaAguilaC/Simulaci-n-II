```python
import numpy as np
import math
```

```python
num=0
prom=0
z=[1.645, 1.663, 1.681, 1.699, 1.71, 1.96, 2.06, 2.08, 2.101, 2.575]
z1=[90, 91, 92, 93, 94, 95, 96, 97, 98, 99]
c=0
con=0
lsp=0
lip=0
lsva=0
liva=0
x1=0
x2=0
pos=0
neg=0
eh=0
vh=0

def indep():
  con=0
  print('¿Cuántos números va a ingresar?')
  num=int(input())
  n=[]
  for i in range (num):
    print('Ingrese el número', i+1, ': ')
    n.append(float(input()))

  print('\n')
  print('¿Cuál es el nivel de confianza? ')
  print('1: 90%')
  print('2: 91%')
  print('3: 92%')
  print('4: 93%')
  print('5: 94%')
  print('6: 95%')
  print('7: 96%')
  print('8: 97%')
  print('9: 98%')
  print('10: 99%')
  c=int(input())
  print('\n')

  cor=[]
  for i in range (num):
    if n[i]<=n[i-1]:
      cor.append(0)
    else:
      cor.append(1)

  cor.pop(0)

  for i in range (num-1):
    if cor[i]!=cor[i-1]:
      con+=1
  con+=1

  eh=((2*num)-1)/3
  vh=((16*num)-29)/90

  z2=0
  z2=(con-eh)/math.sqrt(vh)

  if z2<z[c-1]:
    print('Los', num, 'numeros tienen una distribución uniforme, con un nivel de confianza del', z1[c-1], '%')
  else:
   print('Los', num, 'numeros no tienen una distribución uniforme, con un nivel de confianza del', z1[c-1], '%')

def varianza():
  print('¿Cuántos números va a ingresar?')
  num=int(input())
  n=[]
  for i in range (num):
    print('Ingrese el número', i+1, ': ')
    n.append(float(input()))

  print('\n')
  print('¿Cuál es el nivel de confianza? ')
  print('1: 90%')
  print('2: 91%')
  print('3: 92%')
  print('4: 93%')
  print('5: 94%')
  print('6: 95%')
  print('7: 96%')
  print('8: 97%')
  print('9: 98%')
  print('10: 99%')
  c=int(input())
  print('\n')

  print('Ingrese el valor de x^2(a/2, n-1):')
  x1=float(input())
  print('Ingrese el valor de x^2(1-a/2, n-1):')
  x2=float(input())
  print('\n')
  lsva=x1/(12*(num-1))
  liva=x2/(12*(num-1))

  print('El límite inferior es:', liva)
  print('La varianza es:', np.var(n))
  print('El límite superior es:', lsva)
  print('\n')
  if liva<prom and prom<lsva:
   print('Los', num, 'numeros tienen una varianza estadísticamente igual a 1/12, con un nivel de confianza del', z1[c-1], '%')
  else:
   print('Los', num, 'numeros no tienen una varianza estadísticamente igual a 1/12, con un nivel de confianza del', z1[c-1], '%')

def media():
  suma=0
  print('¿Cuántos números va a ingresar?')
  num=int(input())
  n=[]
  for i in range (num):
    print('Ingrese el número', i+1, ': ')
    n.append(float(input()))

  for i in range (num):
    suma+=n[i]

  prom=suma/num
  print('\n')
  print('¿Cuál es el nivel de confianza? ')
  print('1: 90%')
  print('2: 91%')
  print('3: 92%')
  print('4: 93%')
  print('5: 94%')
  print('6: 95%')
  print('7: 96%')
  print('8: 97%')
  print('9: 98%')
  print('10: 99%')
  c=int(input())
  print('\n')
  lsp=0.5+z[c-1]*(1/(12*math.sqrt(num)))
  lip=0.5-z[c-1]*(1/(12*math.sqrt(num)))

  print('El límite inferior es:', lip)
  print('El promedio es:', prom)
  print('El límite superior es:', lsp)
  print('\n')
  if lip<prom and prom<lsp:
   print('Los', num, 'numeros tienen una media estadísticamente igual a 1/2, con un nivel de confianza del', z1[c-1], '%')
  else:
   print('Los', num, 'numeros no tienen una media estadísticamente igual a 1/2, con un nivel de confianza del', z1[c-1], '%')

print('Elija un método')
print('1. Prueba de medias')
print('2. Prueba de varianza')
print('3. Prueba de independencia')
opc=0
opc=int(input())

if opc == 1:
	media()
elif opc == 2:
	varianza()
elif opc == 3:
	indep()
else:
  print('Elija una opción de la lista')
```
